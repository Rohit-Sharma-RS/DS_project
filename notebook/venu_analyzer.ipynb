```python
venues = df['venue'].unique().tolist()
```

```python
venue_stats = {}

for venue in venues:
    # Filter matches for this venue
    venue_matches = df[df['venue'] == venue]

    # Basic stats
    total_matches = len(venue_matches)

    # Batting first vs fielding first stats
    batting_first_wins = sum((venue_matches['toss_winner'] == venue_matches['winner']) & 
                            (venue_matches['toss_decision'] == 'bat'))
    batting_first_wins += sum((venue_matches['toss_winner'] != venue_matches['winner']) & 
                                (venue_matches['toss_decision'] == 'field'))

    bowling_first_wins = total_matches - batting_first_wins

    # Calculate average scores - first and second innings
    # This requires ball-by-ball data which isn't in this dataset
    # So we'll initialize these values for now
    avg_first_innings = 0
    avg_second_innings = 0

    # If you have additional data with innings scores, you can calculate these:
    # For example, if you had a deliveries or innings dataframe:
    # if 'deliveries_df' in globals():
    #     venue_innings = deliveries_df[deliveries_df['match_id'].isin(venue_matches['id'])]
    #     first_innings = venue_innings[venue_innings['innings'] == 1]
    #     second_innings = venue_innings[venue_innings['innings'] == 2]
    #     avg_first_innings = first_innings.groupby('match_id')['total_runs'].sum().mean()
    #     avg_second_innings = second_innings.groupby('match_id')['total_runs'].sum().mean()

    # Team performance at this venue
    team_stats = {}
    all_teams = set(venue_matches['team1'].tolist() + venue_matches['team2'].tolist())

    for team in all_teams:
        # Matches where team participated
        team_matches = venue_matches[(venue_matches['team1'] == team) | (venue_matches['team2'] == team)]
        matches_played = len(team_matches)
        
        # Matches won by team
        matches_won = len(team_matches[team_matches['winner'] == team])
        
        # Win rate
        win_rate = matches_won / matches_played if matches_played > 0 else 0
        highest_runs = team_matches['target_runs'].max()
        team_stats[team] = {
            'matches_played': int(matches_played),
            'matches_won': int(matches_won),
            'win_rate': round(win_rate * 100, 2)
        }

        # Season-wise match count
        season_matches = venue_matches.groupby('season').size().to_dict()

        # Store all stats for the venue
        venue_stats[venue] = {
            'total_matches': int(total_matches),
            'highest_runs': highest_runs,
            'batting_first_wins': int(batting_first_wins),
            'bowling_first_wins': int(bowling_first_wins),
            'batting_first_win_rate': round((batting_first_wins / total_matches) * 100, 2) if total_matches > 0 else 0,
            'bowling_first_win_rate': round((bowling_first_wins / total_matches) * 100, 2) if total_matches > 0 else 0,
            'team_stats': team_stats,
            'season_matches': {str(k): int(v) for k, v in season_matches.items()},
            'city': venue_matches['city'].iloc[0] if not venue_matches.empty and not pd.isna(venue_matches['city'].iloc[0]) else 'Unknown'
        }
```

```python
matches = df
deliveries = df1
player_elo = pd.read_csv(r"D:\data-science-end\notebook\data\Recent_ELO.csv")

deliveries.head()
```

```python
all_batters = deliveries['batter'].unique().tolist()
all_bowlers = deliveries['bowler'].unique().tolist()
all_non_strikers = deliveries['non_striker'].unique().tolist()
all_players_from_elo = player_elo['Player'].tolist()
```

```python
all_players = list(set([p for p in all_batters + all_bowlers + all_non_strikers + all_players_from_elo if pd.notna(p) and p != 'NA']))
player_analysis = {}
```

```python
for player in all_players:
    # Initialize player stats
    player_stats = {
        "name": player,
        "matches_played": 0,
        "role": "Unknown",  # Will be updated if found in ELO data
        "form": "Unknown",  # Will be updated if found in ELO data
        "last_played": "",  # Will be updated if found in ELO data
        "status": "Unknown",  # Will be updated if found in ELO data
        "batting_stats": {
            "innings": 0,
            "runs": 0,
            "balls_faced": 0,
            "average": 0,
            "strike_rate": 0,
            "highest_score": 0,
            "fifties": 0,
            "hundreds": 0,
            "fours": 0,
            "sixes": 0
        },
        "bowling_stats": {
            "innings": 0,
            "wickets": 0,
            "balls_bowled": 0,
            "runs_conceded": 0,
            "economy": 0,
            "average": 0,
            "best_figures": "0/0"
        },
        "recent_performances": []
    }
    
    player_elo_row = player_elo[player_elo['Player'] == player]
    if not player_elo_row.empty:
        player_stats["role"] = player_elo_row.iloc[0]['Role']
        player_stats["form"] = player_elo_row.iloc[0]['FormCategory']
        player_stats["last_played"] = player_elo_row.iloc[0]['LastPlayed']
        player_stats["status"] = player_elo_row.iloc[0]['Status']
    
    # Calculate batting stats
    player_batting = deliveries[deliveries['batter'] == player]
    if not player_batting.empty:
        player_stats["batting_stats"]["innings"] = len(player_batting['match_id'].unique())
        player_stats["batting_stats"]["runs"] = player_batting['batsman_runs'].sum()
        player_stats["batting_stats"]["balls_faced"] = len(player_batting)
        
        # Calculate strike rate
        if player_stats["batting_stats"]["balls_faced"] > 0:
            player_stats["batting_stats"]["strike_rate"] = round(
                (player_stats["batting_stats"]["runs"] / player_stats["batting_stats"]["balls_faced"]) * 100, 2
            )
        
        # Calculate highest score by grouping by match_id and inning
        if not player_batting.empty:
            player_innings = player_batting.groupby(['match_id', 'inning'])['batsman_runs'].sum()
            if not player_innings.empty:
                player_stats["batting_stats"]["highest_score"] = player_innings.max()
        
        # Count boundaries
        player_stats["batting_stats"]["fours"] = len(player_batting[player_batting['batsman_runs'] == 4])
        player_stats["batting_stats"]["sixes"] = len(player_batting[player_batting['batsman_runs'] == 6])
    
    # Calculate bowling stats
    player_bowling = deliveries[deliveries['bowler'] == player]
    if not player_bowling.empty:
        player_stats["bowling_stats"]["innings"] = len(player_bowling['match_id'].unique())
        player_stats["bowling_stats"]["balls_bowled"] = len(player_bowling)
        player_stats["bowling_stats"]["wickets"] = len(player_bowling[player_bowling['is_wicket'] == 1])
        player_stats["bowling_stats"]["runs_conceded"] = player_bowling['total_runs'].sum() - player_bowling['extra_runs'].sum()
        
        # Calculate economy
        if player_stats["bowling_stats"]["balls_bowled"] > 0:
            overs_bowled = player_stats["bowling_stats"]["balls_bowled"] / 6
            if overs_bowled > 0:
                player_stats["bowling_stats"]["economy"] = round(
                    player_stats["bowling_stats"]["runs_conceded"] / overs_bowled, 2
                )
        
        # Calculate bowling average
        if player_stats["bowling_stats"]["wickets"] > 0:
            player_stats["bowling_stats"]["average"] = round(
                player_stats["bowling_stats"]["runs_conceded"] / player_stats["bowling_stats"]["wickets"], 2
            )
        
        # Calculate best bowling figures by grouping by match_id and inning
        if not player_bowling.empty and player_stats["bowling_stats"]["wickets"] > 0:
            bowling_innings = player_bowling.groupby(['match_id', 'inning'])
            best_wickets = 0
            best_runs = float('inf')
            
            for _, innings_data in bowling_innings:
                wickets = len(innings_data[innings_data['is_wicket'] == 1])
                runs = innings_data['total_runs'].sum() - innings_data['extra_runs'].sum()
                
                if wickets > best_wickets or (wickets == best_wickets and runs < best_runs):
                    best_wickets = wickets
                    best_runs = runs
            
            if best_wickets > 0:
                player_stats["bowling_stats"]["best_figures"] = f"{best_wickets}/{best_runs}"
    
    # Get recent performances
    player_matches = set(player_batting['match_id'].unique()).union(set(player_bowling['match_id'].unique()))
    
    for match_id in list(player_matches)[:5]:  # Last 5 matches
        match_info = matches[matches['id'] == match_id]
        if not match_info.empty:
            match_date = match_info.iloc[0]['date']
            
            # Batting in this match
            match_batting = player_batting[player_batting['match_id'] == match_id]
            batting_runs = match_batting['batsman_runs'].sum() if not match_batting.empty else 0
            
            # Bowling in this match
            match_bowling = player_bowling[player_bowling['match_id'] == match_id]
            bowling_wickets = len(match_bowling[match_bowling['is_wicket'] == 1]) if not match_bowling.empty else 0
            bowling_runs = match_bowling['total_runs'].sum() - match_bowling['extra_runs'].sum() if not match_bowling.empty else 0
            
            performance = {
                "match_id": match_id,
                "date": match_date,
                "batting": {
                    "runs": batting_runs,
                    "balls": len(match_batting) if not match_batting.empty else 0
                },
                "bowling": {
                    "wickets": bowling_wickets,
                    "runs_conceded": bowling_runs,
                    "balls": len(match_bowling) if not match_bowling.empty else 0
                }
            }
            
            player_stats["recent_performances"].append(performance)
    
    # Count matches played
    player_stats["matches_played"] = len(player_matches)
    
    # Add player to analysis
    player_analysis[player] = player_stats
```

```python
for team in all_teams:
        # Initialize team stats
        team_stats = {
            "name": team,
            "matches_played": 0,
            "matches_won": 0,
            "win_percentage": 0,
            "titles": 0,  # Would need additional data
            "batting_stats": {
                "total_runs": 0,
                "highest_score": 0,
                "average_score": 0,
                "run_rate": 0
            },
            "bowling_stats": {
                "total_wickets": 0,
                "economy_rate": 0,
                "best_bowling": "0/0"
            },
            "head_to_head": {},
            "top_batters": [],
            "top_bowlers": [],
            "recent_form": []
        }
        
        # Calculate matches played and won
        team_matches = matches[(matches['team1'] == team) | (matches['team2'] == team)]
        team_stats["matches_played"] = len(team_matches)
        team_stats["matches_won"] = len(team_matches[team_matches['winner'] == team])
        
        # Calculate win percentage
        if team_stats["matches_played"] > 0:
            team_stats["win_percentage"] = round(
                (team_stats["matches_won"] / team_stats["matches_played"]) * 100, 2
            )
        
        # Calculate batting stats (when team was batting)
        team_batting_deliveries = deliveries[deliveries['batting_team'] == team]
        if not team_batting_deliveries.empty:
            team_stats["batting_stats"]["total_runs"] = team_batting_deliveries['total_runs'].sum()
            
            # Calculate highest and average score
            team_innings = team_batting_deliveries.groupby(['match_id', 'inning'])['total_runs'].sum()
            if not team_innings.empty:
                team_stats["batting_stats"]["highest_score"] = team_innings.max()
                team_stats["batting_stats"]["average_score"] = round(team_innings.mean(), 2)
            
            # Calculate run rate
            total_balls = len(team_batting_deliveries)
            if total_balls > 0:
                overs = total_balls / 6
                if overs > 0:
                    team_stats["batting_stats"]["run_rate"] = round(
                        team_stats["batting_stats"]["total_runs"] / overs, 2
                    )
        
        # Calculate bowling stats (when team was bowling)
        team_bowling_deliveries = deliveries[deliveries['bowling_team'] == team]
        if not team_bowling_deliveries.empty:
            team_stats["bowling_stats"]["total_wickets"] = team_bowling_deliveries['is_wicket'].sum()
            
            # Calculate economy rate
            total_runs_conceded = team_bowling_deliveries['total_runs'].sum()
            total_balls = len(team_bowling_deliveries)
            if total_balls > 0:
                overs = total_balls / 6
                if overs > 0:
                    team_stats["bowling_stats"]["economy_rate"] = round(
                        total_runs_conceded / overs, 2
                    )
            
            # Calculate best bowling performance
            team_bowling_innings = team_bowling_deliveries.groupby(['match_id', 'inning'])
            best_wickets = 0
            best_runs = float('inf')
            
            for _, innings_data in team_bowling_innings:
                wickets = innings_data['is_wicket'].sum()
                runs = innings_data['total_runs'].sum()
                
                if wickets > best_wickets or (wickets == best_wickets and runs < best_runs):
                    best_wickets = wickets
                    best_runs = runs
            
            if best_wickets > 0:
                team_stats["bowling_stats"]["best_bowling"] = f"{best_wickets}/{best_runs}"
        
        # Calculate head-to-head
        for opponent in all_teams:
            if opponent != team:
                # Matches where team and opponent played against each other
                h2h_matches = matches[
                    ((matches['team1'] == team) & (matches['team2'] == opponent)) |
                    ((matches['team1'] == opponent) & (matches['team2'] == team))
                ]
                
                if not h2h_matches.empty:
                    wins = len(h2h_matches[h2h_matches['winner'] == team])
                    losses = len(h2h_matches[h2h_matches['winner'] == opponent])
                    no_results = len(h2h_matches) - wins - losses
                    
                    team_stats["head_to_head"][opponent] = {
                        "played": len(h2h_matches),
                        "won": wins,
                        "lost": losses,
                        "no_result": no_results
                    }
        
        # Find top batters for this team
        if not team_batting_deliveries.empty:
            batters = team_batting_deliveries['batter'].unique()
            batter_stats = []
            
            for batter in batters:
                batter_deliveries = team_batting_deliveries[team_batting_deliveries['batter'] == batter]
                runs = batter_deliveries['batsman_runs'].sum()
                balls_faced = len(batter_deliveries)
                
                if balls_faced > 0:
                    strike_rate = round((runs / balls_faced) * 100, 2)
                    
                    batter_stats.append({
                        "name": batter,
                        "runs": runs,
                        "balls_faced": balls_faced,
                        "strike_rate": strike_rate
                    })
            
            # Sort by runs and get top 5
            team_stats["top_batters"] = sorted(batter_stats, key=lambda x: x["runs"], reverse=True)[:5]
        
        # Find top bowlers for this team
        if not team_bowling_deliveries.empty:
            bowlers = team_bowling_deliveries['bowler'].unique()
            bowler_stats = []
            
            for bowler in bowlers:
                bowler_deliveries = team_bowling_deliveries[team_bowling_deliveries['bowler'] == bowler]
                wickets = len(bowler_deliveries[bowler_deliveries['is_wicket'] == 1])
                runs_conceded = bowler_deliveries['total_runs'].sum() - bowler_deliveries['extra_runs'].sum()
                balls_bowled = len(bowler_deliveries)
                
                if balls_bowled > 0:
                    economy = round((runs_conceded / (balls_bowled / 6)), 2)
                    
                    bowler_stats.append({
                        "name": bowler,
                        "wickets": wickets,
                        "runs_conceded": runs_conceded,
                        "balls_bowled": balls_bowled,
                        "economy": economy
                    })
            
            # Sort by wickets and get top 5
            team_stats["top_bowlers"] = sorted(bowler_stats, key=lambda x: x["wickets"], reverse=True)[:5]
        
        # Calculate recent form
        team_matches_sorted = team_matches.sort_values('date', ascending=False)
        recent_matches = team_matches_sorted.head(5)
        
        for _, match in recent_matches.iterrows():
            match_result = "won" if match['winner'] == team else "lost"
            opponent = match['team2'] if match['team1'] == team else match['team1']
            
            form_entry = {
                "match_id": match['id'],
                "date": match['date'],
                "opponent": opponent,
                "result": match_result,
                "margin": match['result_margin'],
                "venue": match['venue']
            }
            
            team_stats["recent_form"].append(form_entry)
        
        # Add team to analysis
        team_analysis[team] = team_stats
```

```python
import numpy as np

# Custom JSON encoder to handle NumPy types
class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.integer):
            return int(obj)
        elif isinstance(obj, np.floating):
            return float(obj)
        elif isinstance(obj, np.ndarray):
            return obj.tolist()
        return super(NumpyEncoder, self).default(obj)

with open('player_analysis.json', 'w') as f:
    json.dump(player_analysis, f, indent=4, cls=NumpyEncoder)

with open('team_analysis.json', 'w') as f:
    json.dump(team_analysis, f, indent=4, cls=NumpyEncoder)
```
